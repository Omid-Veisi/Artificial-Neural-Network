```python
from tensorflow import keras
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout, ReLU
from tensorflow.keras.layers import LeakyReLU
from tensorflow.keras.optimizers import SGD, Adam
import matplotlib.pyplot as plt
import numpy as np
```

**Model 1 from X to y**

```python
# standardize the inputs
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler

X1 = X.copy()
col_names = ['Scale', 'Orientation', 'avg_h']
features = X1[col_names]
scaler1 = StandardScaler().fit(features.values)
features = scaler1.transform(features.values)
X1[col_names] = features

X1
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
y1 = scaler.fit_transform(y)
y1
```

```python
X1_for_model = X1.iloc[:-10]
y1_for_model = y1[:-10]

X1_for_paper = X1.iloc[-10:]
y1_for_paper = y1[-10:]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X1_for_model,y1_for_model,test_size=0.3)
```

```python
model = Sequential()
model.add(Dense(8, input_dim=5, kernel_initializer='he_uniform',  
                 activation=LeakyReLU()))
model.add(Dense(16, activation=LeakyReLU()))
model.add(Dense(32, activation=LeakyReLU()))
model.add(Dense(16, activation=LeakyReLU()))
model.add(Dense(8, activation=LeakyReLU()))
model.add(Dense(3))
model.compile(optimizer=Adam(0.001),loss = 'mse')
```

```python
for i in range(20):
  model.fit(X_train, y_train,validation_data=(X_test, y_test),\
            epochs=10, verbose=2)
  y_pred = model.predict(X_test)
  print (f'loop {i} : r2score = {r2_score(y_test,y_pred)}')
  print ('-'*30)
```
